# Matrix Functions

Transpose function

```python
a = np.matrix([[4,5,6],[14,16,18],[9,17,9]])
print(a)
print()
print(np.transpose(a))
```

```python
a = np.matrix([[4,5,6],[14,16,18],[9,17,9]])
print(a)
print()
print(a.T)
```

Swapaxes function

```python
a = np.matrix([[4,5,6],[14,16,18],[9,17,9]])
print(a)
print()
print(np.swapaxes(a,1,0))
```

```python
a = np.matrix([[4,5,6],[14,16,18],[9,17,9]])
print(a)
print()
print(np.swapaxes(a,0,1))
```

Inverse

```python
a = np.matrix([[3,4],[2,1]])
print(a)
print()
print(np.linalg.inv(a))
```

Power

```python
a = np.matrix([[1,2],[4,5]])
print(a)
print()
print(np.linalg.matrix_power(a,2))
print()
print(np.linalg.matrix_power(a,0))
print()
print(np.linalg.matrix_power(a,-2))
```

Determinate

```python
a = np.matrix([[1,3,5],[2,4,6],[10,11,12]])
print(a)
print()
print(np.linalg.det(a))
```
