# Data Types in numpy

```python
import numpy as np
```

```python
x = np.array([1,2,3,4])
print("The data type is:", x.dtype)
```

```python
x1 = np.array([2.5,3.5,3.7, 6.7,5.8,4.9])
print("The data type is:", x1.dtype)
```

```python
x2 = np.array(["apple", "dog", "horse", "cat"])
print("The data type is:", x2.dtype)
```

```python
x3 = np.array(["apple", "dog", "horse", "cat", 4,5,6,7,3])
print("The data type is:", x3.dtype)
```

# converting one data type into another data type

```python
x = np.array([1,2,3,4,5])
print(x)
print(x.dtype)

x1 = np.array([1,2,3,4,5], dtype=np.int16)
print(x1)
print(x1.dtype)
```

```python
x = np.array([1,2,3,4,5])
print(x)
print(x.dtype)

x1 = np.array([1,2,3,4,5], dtype="f")
print(x1)
print(x1.dtype)
```

```python
m = np.array([1,2,3,4])
print(m)
print(m.dtype)

m1 = np.float64(m)
print(m1)
print(m1.dtype)
```

```python
m = np.array([1,2,3,4])
print(m)
print(m.dtype)

m1 = np.float64(m)
print(m1)
print(m1.dtype)

m2 = np.int64(m1)
print(m2)
print(m2.dtype)
```

```python
x = np.array([4.5,5.7,6.7,7.3,8.7])
y = x.astype(complex)
print(x)
print(x.dtype)
print(y)
print(y.dtype)
```

# Arithmetic Operations in numpy array

Addition

```python
x = np.array([1,2,3,4,5])
y = x+3
print(y)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
z = x+y
print(z)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
z = np.add(x,y)
print(z)
```

Subtraction

```python
x = np.array([1,2,3,4,5])
y = x-2
print(y)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
z = y-x
print(z)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
z = np.subtract(x,y)
print(z)
```

Multiplication

```python
x = np.array([1,2,3,4,5])
y = x*2
print(y)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
z = y*x
print(z)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
z = np.multiply(x,y)
print(z)
```

Division

```python
x = np.array([1,2,3,4,5])
y = x/3
print(y)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
z = y/x
print(z)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])
z = np.divide(x,y)
print(z)
```

Modulous

```python
x = np.array([10,20,30,40,50])
y = x%3
print(y)
```

```python
x = np.array([16,12,43,45,54])
y = np.array([5,6,7,8,9])
z = y%x
print(z)
```

```python
x = np.array([16,23,33,45,54])
y = np.array([5,6,7,8,9])
z = np.mod(x,y)
print(z)
```

Reciprocal

```python
a = np.array([23,25,28,30])
b = np.reciprocal(a)
print(b)
```

# Arithmentic Operations on 3d arrays

```python
n1 = np.array([[[1,2,3,4],[5,6,7,8]]])
n2 = np.array([[[1,2,3,4],[5,6,7,8]]])
x = n1+n2
print(x)
print(x.ndim)
```

```python
n1 = np.array([[[1,2,3,4],[5,6,7,8]]])
n2 = np.array([[[1,2,3,4],[5,6,7,8]]])
x = n1-n2
print(x)
print(x.ndim)
```

```python
n1 = np.array([[[1,2,3,4],[5,6,7,8]]])
n2 = np.array([[[1,2,3,4],[5,6,7,8]]])
x = n1*n2
print(x)
print(x.ndim)
```

```python
n1 = np.array([[[1,2,3,4],[5,6,7,8]]])
n2 = np.array([[[1,2,3,4],[5,6,7,8]]])
x = n1/n2
print(x)
print(x.ndim)
```

# Arithmetic Functions

max & min along with index number

```python
a = np.array([12,13,14,15,16,17,56,87,83])
print(np.max(a))
print(np.min(a))
```

```python
a = np.array([12,13,14,15,16,17,56,87,83])
print(np.max(a), np.argmax(a))
print(np.min(a), np.argmin(a))
```

```python
m = np.array([[34,24,45,67,68],[32,49,21,37,12]])
print(m)
print(m.ndim)
n = np.min(m, axis=0) # 0 - row
print(n)
```

```python
m = np.array([[34,24,45,67,68],[32,49,21,37,12]])
print(m)
print(m.ndim)
n = np.min(m, axis=1) # 1 - col
print(n)
```

```python
m = np.array([[34,24,45,67,68],[32,49,21,37,12]])
print(m)
print(m.ndim)
n = np.max(m, axis=0) # 0 - row
print(n)
```

```python
m = np.array([[34,24,45,67,68],[32,49,21,37,12]])
print(m)
print(m.ndim)
n = np.max(m, axis=1) # 1 - col
print(n)
```
